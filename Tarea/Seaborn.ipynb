```python
import pandas as pd

def read_csv_to_dataframe(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except Exception as a:
        print("Error reading the CSV file: {a}")
        return None
```

```python
terminos_df = read_csv_to_dataframe("Terminos_lagoon_TA_DIC_2023_RawData.csv")
```

```python
print(terminos_df.info())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="ticks", palette="bright") 

```

```python
# Load the example tips dataset
ax = sns.catplot (data=terminos_df,
                 x="area", y="ta_micromol_kg",
                 hue="season", col="estuary",
                 kind="box",
                 height=5, aspect=1)

ax.set_titles("{col_name}")

for a in ax.axes.flat:
    a.set_title(a.get_title(), fontsize=18, fontweight="bold")

plt.show()

```
